```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
print(dataset.DESCR)
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
dataset.target
```

```python
## Independent and dependent
X = df.iloc[:,:]
y = dataset.target
```

```python
X
```

```python
y.shape
```

```python
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
import random

# Determine the number of samples to match
min_samples = min(X.shape[0], y.shape[0])

# Randomly sample the larger variable
random.seed(123)  # Set the random seed for reproducibility
if X.shape[0] > y.shape[0]:
    X = X.sample(n=min_samples, random_state=123)
else:
    y = y.sample(n=min_samples, random_state=123)
    
```

```python
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
import random

# Set the random seed for reproducibility
random.seed(123)
# Split the dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## Post Pruning
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree
plt.figsize = (15,10)
tree.plot_tree(treeclassifier,filled = True)
```

```python
##  Prediction
y_pred = treeclassifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```
